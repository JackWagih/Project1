# Banking System using OOP

```python
class User():
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    
    def show_details(self):
        print(f'user name is {self.name}')
        print(f'user age is {self.age} years old')
        print(f'the user is {self.gender}')
        
class Bank(User):
    def __init__(self,name,age,gender):
        User.__init__(self,name,age,gender)
        self.balance=0
        
    def deposit(self,amount):
        self.balance +=amount
        print(f'balance is updated to be {self.balance}')
        
    def withdraw(self,amount):
        if self.balance < amount:
            print('Insufficient amount')
        else:
            self.balance -= amount
            print('balance is updated to be ',self.balance)
            
    def view_balance(self):
        print('The balance now is: ',self.balance)
```

# Main Function

```python
name=input('please enter your name: ')
while True:
    try:
        age=int(input('please enter your age: '))
    except ValueError:
        print('Invalid,please input a number')
        continue
    else:
        break
gender=input('please write your gender: ')    
bank=Bank(name,age,gender)
print ('account is created successfully\n your name is:',name,'\n your age is:',age,'\n you are:',gender)

while True:
    print("""-----Menu-----
            1. show account details
            2. show account balance
            3. withdraw
            4. deposit
            0. Exit
            """)
    try:
        number=int(input('Please enter the number:'))
    except ValueError:
        print('Invalid,please input a number')
        continue
    else:
        if number == 1:
            bank.show_details()
        elif number == 2:
            bank.view_balance()
        elif number == 3:
            amount=float(input('please enter the amount:'))
            bank.withdraw(amount)
        elif number == 4:
            amount=float(input('please enter the amount'))
            bank.deposit(amount)
        elif number == 0:
            print('Thanks for using our banking system')
        else:
            print('Incorrect choice,please enter a number from 0 to 4')
```
